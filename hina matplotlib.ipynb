# matplotlib

```python
import matplotlib.pyplot as plt
```

```python
names=["a","b","c","d"]
marks=[98,88,91,89]
att=[60,57,41,77]
plt.plot(names,marks)
```

```python
plt.plot(names,att)
```

```python
plt.plot(names,marks)
plt.plot(names,att)
```

```python
plt.plot(names,marks,"oy",names,att,"--r")
```

```python
plt.plot(names,marks,"oy",names,att,"--r")
plt.grid()
```

```python
plt.plot(names,marks,"oy",names,att,"--r")
plt.grid()
plt.title("students details")
```

```python
plt.plot(names,marks,"oy",names,att,"--r")
plt.grid()
plt.title("students details")
plt.xlabel("names")
plt.ylabel("details")
```

```python
plt.plot(names,marks,"oy",names,att,"--r")
plt.grid()
plt.title("students details")
plt.xlabel("names")
plt.ylabel("details")
plt.legend(["marks","att"],loc="upper right")
```

```python
plt.subplot(221)
plt.plot(names,att,"og")
plt.subplot(223)
plt.plot(names,att,"--r")
plt.subplot(224)
plt.plot(names,att,"*m")
```
